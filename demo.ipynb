###### Read namelist.txt

```python
import ACPNamelist
fname_namelist = r'C:\Users\GEVENSON\GitHub\simMgmtTrajectory\namelist.txt'
acp_namelist = ACPNamelist.ACPNamelist(fname_namelist)
```

###### Read spatial data and visualize

```python
import ACPDataSpatial
acp_data_spatial = ACPDataSpatial.ACPDataSpatial(acp_namelist)
domain_map = acp_data_spatial.get_folium_map()
display(domain_map)
```

###### Read nutrient yield data

```python
import ACPDataNutrientYields
acp_data_nutrient_yields = ACPDataNutrientYields.ACPDataNutrientYields(acp_namelist,acp_data_spatial,method='swat pdf')
```

###### Read effectiveness coefficient data

```python
import ACPDataEffectivenessCoefficients
acp_data_ecs = ACPDataEffectivenessCoefficients.ACPDataECs(acp_namelist,acp_data_spatial)
```

###### Read cost data

```python
import ACPDataCosts
acp_data_costs = ACPDataCosts.ACPDataCosts(acp_namelist,acp_data_spatial)
```

###### Simulate a set of ACP management scenarios

```python
import ACPScenario
acpscenarioset = ACPScenario.ACPScenarioSet(name='test',acpnamelist=acp_namelist,acpspatial=acp_data_spatial,baseline_data=acp_data_nutrient_yields,ec_data=acp_data_ecs,cost_data=acp_data_costs,n_scenarios=20)
```

###### Visualize results

```python
fig1 = acpscenarioset.graph_delta_tn_tp(acpspatial=acp_data_spatial,baseline_data=acp_data_nutrient_yields,x_var='cost',y_var='% mean annual load',basin_out_rch_num=42)
```

```python
fig2 = acpscenarioset.graph_delta_tn_tp(acpspatial=acp_data_spatial,baseline_data=acp_data_nutrient_yields,x_var='treated area',y_var='% mean annual load',basin_out_rch_num=42)
```

```python
fig3 = acpscenarioset.graph_delta_tn_tp(acpspatial=acp_data_spatial,baseline_data=acp_data_nutrient_yields,x_var='count',y_var='% mean annual load',basin_out_rch_num=42)
```
